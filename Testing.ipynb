# Plotting Numerical Range of a $2 \times 2$ Matrix

Consider the matrix
$$ A = \begin{pmatrix}-1 & 0\\ 5 & -1 + 5i\end{pmatrix}$$

The numerical range of a $2 \times 2$ matrix is an ellipse, possibly degenerate.

```python
from ellipses import getOneEllipse, getEllipses, makePatch
from plotting import plotEllipses
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.collections import PatchCollection
```

```python
# Define Matrix
j = complex(0,1)
A = [[-1,0],[5,-1+5*j]]

# Find numerical range of matrix
e = getOneEllipse(A)

# [(center x, center y), major axis length, minor axis length, angle]
e
```

```python
# create ellipse patch through matplotlib.patches
el = makePatch(A)
print(el.center)
print(el.width)
print(el.height)
print(el.angle)
```

```python
plotEllipses([el])
plt.savefig("test.png",dpi=300)
```

# Plotting Numerical Range of an Operator

```python
j = complex(0,1)
A0 = [[-1,0],[0,-1 + 5*j]]
A0Mat = np.array(A0)
A1 = [[-j - 2,0],[0,j + 1]]
A1Mat = np.array(A1)

els = getEllipses(A0Mat,A1Mat,100)
plotEllipses(els)
```
