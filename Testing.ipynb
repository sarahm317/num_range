# Plotting Numerical Range of a $2 \times 2$ Matrix

Consider the matrix
$$ A = \begin{pmatrix}-1 & 0\\ 5 & -1 + 5i\end{pmatrix}$$

The numerical range of a $2 \times 2$ matrix is an ellipse, possibly degenerate.

```python
from ePlotter import getOneEllipse, getEllipses, plotEllipses
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Define Matrix
j = complex(0,1)
A = [[-1,0],[5,-1+5*j]]

# Find numerical range of matrix
e = getOneEllipse(A)

# Plot numerical range
plotEllipses([e])
```

# Plotting Numerical Range of an Operator

```python
j = complex(0,1)
A0 = [[-1,0],[0,-1 + 5*j]]
A0Mat = np.array(A0)
A1 = [[-j - 2,0],[0,j + 1]]
A1Mat = np.array(A1)

els = getEllipses(A0Mat,A1Mat,3)
e = els[0]
plotEllipses([e])
```

```python
from matplotlib.collections import PatchCollection

els = getEllipses(A0Mat,A1Mat,1)
fig, ax = plt.subplots(subplot_kw={'aspect': 'equal'})
plt.rcParams.update({'font.size': 18})
collection = PatchCollection(els, cmap=plt.cm.hsv, alpha=0.3)
ax.add_collection(collection)
```
